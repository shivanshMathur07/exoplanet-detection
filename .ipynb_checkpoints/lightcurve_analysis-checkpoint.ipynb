```python
#getting all the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#data file
file_txt = pd.read_csv('kepler5-b_transit.txt')
file_txt.to_csv('data_transit.csv' , index=None)
```

```python
data = pd.read_csv('data_transit.csv' , header=None)
data.columns = ["Data"]
data
```

#data cleaning and header removing

```python
#spliting the data
data[["HDJ" , "Norm_Flux"]] = data.Data.str.split(expand = True)
data = data.drop('Data' , 1)
data
```

```python
#cleaning other values
data.isna().sum()
```

```python
data.dtypes
```

```python
#convert the values into float
data['HDJ'] = data['HDJ'].astype(str).astype(float)
data['Norm_Flux'] = data['Norm_Flux'].astype(str).astype(float)
data.dtypes
```

Plotting Lightcurves

```python
fig = px.scatter(data , x="HDJ" , y="Norm_Flux",title="Kepler 5-b Lightcurve")
fig.update(layout = dict(title = dict(x=0.5)))
# fig.show()
```
