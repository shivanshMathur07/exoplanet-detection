```python
selected = data
print(selected.shape)
selected = selected.loc[:,selected.columns.intersection(['ExoplanetArchiveDisposition','DispositionUsingKeplerData','OrbitalPeriod','OrbitalPeriodUpperUnc','OrbitalPeriodLowerUnc','TransitEpoch','TransitEpochUpperUnc',
                                                        'TransitEpochLowerUnc','ImpactParamete','ImpactParameterUpperUnc','ImpactParameterLowerUnc',
                                                        'TransitDuration','TransitDurationUpperUnc','TransitDurationLowerUnc','TransitDepth','TransitDepthLowerUnc','TransitDepthUpperUnc',
                                                        'PlanetaryRadius','PlanetaryRadiusUpperUnc','PlanetaryRadiusLowerUnc','EquilibriumTemperature','InsolationFlux','InsolationFluxUpperUnc',
                                                        'InsolationFluxLowerUnc','TransitSignal-to-Nois','TCEPlanetNumbe','StellarEffectiveTemperature','StellarEffectiveTemperatureUpperUnc',
                                                        'StellarEffectiveTemperatureLowerUnc','StellarSurfaceGravity','StellarSurfaceGravityUpperUnc','StellarSurfaceGravityLowerUnc','StellarRadius',
                                                        'StellarRadiusUpperUnc','StellarRadiusLowerUnc','RA','Dec','Kepler-band'])]
print(selected.shape)
selected.head()
```

```python
#target variables
selected['ExoplanetCandidate'] = selected['DispositionUsingKeplerData'].apply(lambda x: 1 if x == 'CANDIDATE' else 0)
selected['ExoplanetConfirmed'] = selected['ExoplanetArchiveDisposition'].apply(lambda x: 2 if x == 'CONFIRMED' else 1 if x == 'CANDIDATE' else 0 )
#selected.drop(columns=['DispositionUsingKeplerData','ExoplanetArchiveDisposition'],inplace=True)
print(selected.shape)
selected.head()
```

```python
selected.drop(columns=['DispositionUsingKeplerData','ExoplanetArchiveDisposition'],inplace=True)
```

```python
#checking for empty values
selected.isna().sum()
selected=selected.dropna(axis=0)
selected.shape
```

```python
#checking for infinity values
infy_vals = np.isinf(selected).values.sum()
infy_vals
```

```python
##Performing normalization of data
from sklearn.preprocessing import MinMaxScaler
new_data = selected.drop(columns=['ExoplanetCandidate','ExoplanetConfirmed'])
normalized = MinMaxScaler().fit(new_data)
norm_data = normalized.transform(new_data)
norm_data = pd.DataFrame(norm_data)
#have to add back the target variables back to data
norm_data['ExoplanetCandidate'] , norm_data['ExoplanetConfirmed'] = selected['ExoplanetCandidate'] , selected['ExoplanetConfirmed']
norm_data
```

```python
##Performing standardization of data
from sklearn.preprocessing import StandardScaler
stand_data = selected.copy()
s_cols = ['OrbitalPeriod','OrbitalPeriodUpperUnc','OrbitalPeriodLowerUnc','TransitEpoch','TransitEpochUpperUnc',
        'TransitEpochLowerUnc','ImpactParamete','ImpactParameterUpperUnc','ImpactParameterLowerUnc',
        'TransitDuration','TransitDurationUpperUnc','TransitDurationLowerUnc','TransitDepth','TransitDepthLowerUnc','TransitDepthUpperUnc',
        'PlanetaryRadius','PlanetaryRadiusUpperUnc','PlanetaryRadiusLowerUnc','EquilibriumTemperature','InsolationFlux','InsolationFluxUpperUnc',
        'InsolationFluxLowerUnc','TransitSignal-to-Nois','TCEPlanetNumbe','StellarEffectiveTemperature','StellarEffectiveTemperatureUpperUnc',
        'StellarEffectiveTemperatureLowerUnc','StellarSurfaceGravity','StellarSurfaceGravityUpperUnc','StellarSurfaceGravityLowerUnc','StellarRadius',
         'StellarRadiusUpperUnc','StellarRadiusLowerUnc','RA','Dec','Kepler-band']
for i in s_cols:
    standard = StandardScaler().fit(stand_data[[i]])
    stand_data[i] = standard.transform(stand_data[[i]])
stand_data
# selected
```

```python
##Feature & Target Extraction
#for main data
features_main = selected.drop(columns=['ExoplanetCandidate','ExoplanetConfirmed'])
target_main = selected.ExoplanetConfirmed

#for normalised data
features_norm = norm_data.drop(columns=['ExoplanetCandidate','ExoplanetConfirmed'])
target_norm = norm_data.ExoplanetConfirmed

#for standardised data
features_stand = stand_data.drop(columns = ['ExoplanetCandidate','ExoplanetConfirmed'])
target_stand = stand_data.ExoplanetConfirmed
```

```python
##splitting training and testing data
from sklearn.model_selection import train_test_split
#for main data
features_main_train , features_main_test , target_main_train , target_main_test = train_test_split(features_main , target_main , test_size=0.3 )
print(features_main_train.shape , features_main_test.shape , target_main_train.shape, target_main_test.shape)

#for normalised data
features_norm_train , features_norm_test , target_norm_train , target_norm_test = train_test_split(features_norm , target_norm , test_size=0.3 )
print(features_norm_train.shape , features_norm_test.shape , target_norm_train.shape, target_norm_test.shape)

#for standardised data
features_stand_train , features_stand_test , target_stand_train , target_stand_test = train_test_split(features_stand , target_stand , test_size=0.3 )
print(features_stand_train.shape , features_stand_test.shape , target_stand_train.shape, target_stand_test.shape)
```

```python
##model evaluation
##accuracy,precision,recall,f1 score
from sklearn.metrics import accuracy_score , recall_score, precision_score, f1_score, confusion_matrix
def evaluate_score(true,pred):
    print('Accuracy Score is ' + accuracy_score(true , pred))
    print('Accuracy Score is ' + recall_score(true , pred))
    print('Accuracy Score is ' + precision_score(true , pred))
    print('Accuracy Score is ' + f1_score(true , pred))
    print('Accuracy Score is ' + confusion_matrix(true , pred))
```
